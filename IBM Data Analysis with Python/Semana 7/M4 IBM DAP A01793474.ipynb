Modelo - Equación matematica para predecir un valor cuando se nos provee de uno o más valores. Esto a travez de la relación entre variables dependientes e independientes. Entre mayor información relevante mejor nuestro modelo.

    
    Regresion Lineal Simple: Se refiere a utilizar solo una variable independiente para realizar una predicción de datos. Nos ayuda a ver la relación entre dos variables.
        
        - Predictor: X variable independiente.
        - Meta (Target): Y variable dependiente.

        y = b0 + b1 X

        b0 = intercept
        b1 = slope

    Regresión Lineal Multiple: Se refiere a utilizar más de una variable independiente para realizar una predicción. Nos ayuda a obtener predicciones con más variables para obtener una variable de salida.

        - Meta (Target): variable CONTINUA Y.
        - Predictores (2 o más): variables X.

        y = b0 + b1x1 + b2x2 + .. + bnxn

        b0 = intercept
        b1 = coeficiente of x1
        ...
        bn = coeficiente of xn

```python
# Simple Lineal Model Estimator

from sklearn.linear_model import LinearRegression

# Constructor
lm=LinearRegression()

# Define Predictor and Target
X = df[["highway-mpg"]]
Y = df[["price"]]

# Fit the model
lm.fit(X,Y)

# Get a prediction
Yhat=lm.predict(X)

# Prediction of the values X
print(Yhat)

# Show intercept value (b0)
print(lm.intercept_)

# Show slope value (b1)
print(lm.coef_)
```

```python
# Multiple Linear Model Estimator

mlm=LinearRegression()
Z = df[["horsepower","curb-weight","engine-size","highway-mpg"]]

# Fit Model
mlm.fit(Z,Y)

# Make a prediction
Yhat2=mlm.predict(Z)

# Prediction of the values X
print(Yhat2)

# Show intercept value (b0)
print(mlm.intercept_)

# Show slopes values (b1..bn)
print(mlm.coef_)
```

Grafica de Residuales
- Se grafican en un punto las diferencias entre predicción y real ( Yhat - Y ).
- X es el valor de la variable dependiente predictor.
- Y es el residuo de la diferencia entre Yhat y Y.
- Se espera que el promedio este en el centro de la grafica y que la varianza se equitativa a una normal.
- Si la grafica muestra una distribución no aleatoria es posible que el modelo no debe ser lineal.
- Si la varianza aumenta con la variable x significa que nuestro modelo no es correcto.
    

```python
# Residual Plot
import seaborn as sns

# x = independant variable predictor
# y = dependant variable target

sns.residplot(x=df["highway-mpg"], y=df["price"])
```

Regressión Polinomial
- Es un caso especial de las regresiones lineales.
- Describe relaciones con curvas valores de predictores elevados a la n potencias.

        - Meta (Target): variable CONTINUA Y.
        - Predictor: variables X a la n potencia

        y = b0 + b1x1 + b2(x)^2 + .. + bn(x)^n

        b0 = intercept
        b1 = coeficiente of x1
        ...
        bn = coeficiente of x^n



```python
# Polinomial Regression 1d

# Calculate Polinomial regression
f=np.polyfit(df["highway-mpg"],df["price"],3)
p=np.poly1d(f)

print(p)
```

```python
# Pipelines 
# Create Steps to a prediction
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

# Create the steps 
input=[("scale",StandardScaler()),
        ("polynomial",PolynomialFeatures(degree=2)),
        ("model",LinearRegression())
    ]

# Construct with steps
pipe=Pipeline(input)

# Train Pipeline
pipe.fit(df[["horsepower","curb-weight","engine-size","highway-mpg"]],df["price"])
yhat=pipe.predict(df[["horsepower","curb-weight","engine-size","highway-mpg"]])
yhat
```

Metricas de evaluación de modelos predictivos
- Promedio de los errores al cuadrado (MSE)
    
    MSE = ( SUMA ( Yhat - y ) ^ 2 ) / n_datos

- R^2 o Coeficiente deterministico: que tan cerca estan los datos predecidos contra los datos reales. Utiliza el porcentaje de variación de los datos. Si esta cerca de 1 significa que el modelo es un buen predictor de los datos. Si es negativo es que esta sobreentrenado.

    R^2 = ( 1 - (MSE regresión lineal/ MSE promedio de los datos ) )

```python
# Mean Squared Error
from sklearn.metrics import mean_squared_error

mean_squared_error(df["price"],yhat)
```

```python
# R^2
from sklearn.metrics import r2_score

# 1
lm.score(X,Y)

# 2
r_squared = r2_score(Y, p(X))
print('The R-square value is: ', r_squared)
```

```python
# Function to plot polynomial data
import matplotlib.pyplot as plt

def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```
